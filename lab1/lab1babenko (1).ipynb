Задание 1.

Нужно подобрать такую скорость обучения и количество итераций, чтобы наступила сходимость.

```python
import numpy as np

X = np.array([[ 1,  1],
              [ 1,  1],
              [ 1,  2],
              [ 1,  5],
              [ 1,  3],
              [ 1,  0],
              [ 1,  5],
              [ 1, 10],
              [ 1,  1],
              [ 1,  2]])

y = [45, 55, 50, 55, 60, 35, 75, 80, 50, 60]

# функция для вычислиния среднеквадратичного отклонения
def calc_mse(y, y_pred):
    err = np.mean((y - y_pred)**2)
    return err

n = X.shape[0]
```

В нашем случае оптимальным шагом eta = 0.01 (1e-1) и количество итерараций, при которых мы достигаем сходимости == 200

```python
eta = 1e-1
n_iter = 200

W = np.array([1, 0.5])

print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

for i in range(n_iter):
    y_pred = np.dot(X, W) 
    err = calc_mse(y, y_pred)
    for k in range(W.shape[0]): 
        W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))
    if i % 10 == 0:
        eta /= 1.1
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err, 2)}')
```

Задание 3. Нам нужно изменить алгоритм так, чтобы он останавливался, когда веса перестают изменяться меньше определённого значения $epsilon$. Для этого нам понадобиться формулы: 

$w^{k} = w^{k-1} - \eta_{k}\nabla Q(w^{k-1}, X)$ 

 $||w^{k} - w^{k-1}|| < \epsilon$ 
 
 Из которых мы выводим что:

 $\eta_{k}\nabla Q(w^{k-1}, X) < \epsilon$

 Закодим, и зададим значение epsilon == 1

```python
epsilon = 0.1

W = np.array([1, 0.5])

print(f'Number of objects = {n} \
       \nInitial weights = {W} \n')
i = 1
while True:
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)
#     for k in range(W.shape[0]):
#         W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))
    # ИЗМЕНЕНИЯ
    # print(eta *( 2 / n * X.T @ ((X @ W) - y)))
    W -= eta * ( 2 / n * X.T @ ((X @ W) - y))
    if np.linalg.norm((2 / n * X.T @ ((X @ W) - y))) < epsilon:
        break
    # ИЗМЕНЕНИЯ
    #
    if i % 10 == 0:
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err,2)}')
    i += 1
```

Это значение слишком большое, и при нём сходимости мы не наблюдаем. Установим epsilon == 0.01.

```python
epsilon = 0.01

W = np.array([1, 0.5])

print(f'Number of objects = {n} \
       \nInitial weights = {W} \n')
i = 1

# создаём бесконечный цикл
while True:
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)

    W -= eta * ( 2 / n * X.T @ ((X @ W) - y))

    # находим норму вектора, если значение будет меньше епсилон - останавливаем цикл 
    if np.linalg.norm((2 / n * X.T @ ((X @ W) - y))) < epsilon:
        break

    # ИЗМЕНЕНИЯ
    #
    if i % 10 == 0:
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err,2)}')
    i += 1
```

При таком, и меньших значениях epsilon, мы будем наблюдать сходимость!
